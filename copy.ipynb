```python
from agents import Agent104, Agent156, RandomAgent
from env.game import Bigtwo


game = Bigtwo()
players = [
    Agent156("cuda:0"),
    # Agent104("cuda:0"),
    # Agent156("cuda:0"),
    RandomAgent(),
    RandomAgent(),
    RandomAgent(),
]

for _ in range(1000):
    game_num = 100
    game.reset()
    winners = [0, 0, 0, 0]
    losses = [0, 0, 0, 0]
    for i in range(game_num):
        while game.winner == None:
            index = game.player_to_act
            players[index].act(game)

        for i, player in enumerate(players):
            if player == game.winner:
                player.rewards.append(1)
            else:
                player.rewards.append(-1)
        winners[game.winner] += 1
        game.reset()

    for i, player in enumerate(players):
        loss = player.learn()
        losses[i] += loss

    print("Record:", winners)
    print("Losses:", losses)
```
